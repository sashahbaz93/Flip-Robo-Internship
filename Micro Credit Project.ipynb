# DATA Visualisation

Our feature variable of interest is Micro Credit analysis.

It refers to the probability for each loan transaction, whether the customer will be paying back the loaned amount within 5 days of insurance of loan. 

In this case, Label ‘1’ indicates that the loan has been payed i.e. Non- defaulter, while, Label ‘0’ indicates that the loan has not been payed i.e. defaulter.  

# Univariate analysis

```python
plt.subplots(figsize=(8,6))
sns.countplot(x="label", data=df)
plt.title("Micro Credit Analysis")
plt.xlabel('1= non defaulter and 0= defaulter')
plt.ylabel("Credit analysis count")
plt.show()
```

Hence , its observed that pretty much most of the loan takers were not defaulters. Only few stands to be the defaulters. 

```python
plt.subplots(figsize=(6,4))
sns.countplot(x="payback30", data=df)
plt.title("Average payback time in days over last 30 days")
plt.show()
```

```python
plt.subplots(figsize=(6,4))
sns.countplot(x="payback90", data=df)
plt.title("Average payback time in days over last 90 days")
plt.show()
```

```python
plt.subplots(figsize=(6,4))
sns.countplot(x="cnt_ma_rech30", data=df)
plt.title("Number of times main account got recharged in last 30 days")
plt.show()
```

```python
#Observation: Graph shows the variations in account got recharged in last 30 days. Maximum amount count was around 36000.
```

```python
plt.subplots(figsize=(6,4))
sns.countplot(x="sumamnt_ma_rech30", data=df)
plt.title("Total amount of recharge in main account over last 30 days (in Indonesian Rupiah)")
plt.show()
```

```python
df['cnt_da_rech30'].hist()
plt.xlabel('cnt_da_rech30')
plt.ylabel('count')
plt.title("Number of times data account got recharged in last 30 days")
plt.show()
```
